```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
size = 28
raw_data= np.loadtxt('pa2_test_no_label.csv',delimiter=',',skiprows=1)

digit_data = []
for i in range(len(raw_data)):
    digit_data.append(raw_data[i,0:785])

digit_data.sort(key=lambda x: x[0]) # sort array by label
```

```python
# draw digit images
plt.figure(figsize=(15, 60))

print(len(digit_data))

for i in range(len(digit_data)):
    X, Y = np.meshgrid(range(size),range(size))
    Z = digit_data[i].reshape(size,size)   # convert from vector to 28x28 matrix
    Z = Z[::-1,:]             # flip vertical
    plt.subplot(82, 20, i+1)  # layout 200 cells
    plt.xlim(0,27)
    plt.ylim(0,27)
    plt.pcolor(X, Y, Z)
    plt.flag()
    plt.gray()
    plt.tick_params(labelbottom="off")
    plt.tick_params(labelleft="off")

plt.show()
```

```python
size = 28
raw_data = np.loadtxt('pa2_train.csv', delimiter=',', skiprows=1)

digit_data_3 = raw_data[4, 1:785]
digit_data_5 = raw_data[0, 1:785]

plt.figure()
x, y = np.meshgrid(range(size), range(size))
z1 = digit_data_3.reshape(size, size)
z1 = z1[::-1, :]
z2 = digit_data_5.reshape(size, size)
z2 = z2[::-1, :]

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.pcolor(x, y, z1)
ax2.pcolor(x, y, z2)
plt.flag()
plt.gray()

ax1.tick_params(labelbottom="off")
ax1.tick_params(labelleft="off")
ax2.tick_params(labelbottom="off")
ax2.tick_params(labelleft="off")
```
